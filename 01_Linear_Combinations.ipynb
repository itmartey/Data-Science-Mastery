```python
import numpy as np
import matplotlib.pyplot as plt

# Define our 'Basis' vectors (The columns of Matrix A)
v1 = np.array([2, 1])
v2 = np.array([-1, 1])

# Define our 'Target' vector (The result b in Ax=b)
# Let's say we want to reach x=1, y=5
target = np.array([1, 5])

# Solve for weights (x) using NumPy (The 'Reverse' calculation)
# In math: Ax = b  =>  x = A^-1 * b
A = np.column_stack((v1, v2))
weights = np.linalg.solve(A, target)

print(f"To reach {target}, we need {weights[0]} of v1 and {weights[1]} of v2.")

# Visualization
origin = np.array([[0, 0], [0, 0]])
plt.quiver(*origin, A[0,:], A[1,:], color=['r', 'b'], scale=10, label='Columns of A')
plt.quiver(0, 0, target[0], target[1], color='g', scale=10, label='Target (b)')
plt.grid()
plt.legend()
plt.show()
```
